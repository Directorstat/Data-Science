```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv_imshow
```

```python
#Import Image
img = cv2.imread('/content/Corrupted Student Picture.png')
```

```python
#Too check the dimension of the image
img.shape
```

```python
#To view the image

plt.imshow(img)
```

```python
cv_imshow(img)
```

```python
#gray scaling
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv_imshow(gray_image)
```

```python
# rgb image
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv_imshow(rgb_image)
```

```python
# contrast / brightness control

alpha = 3
beta = 0.5

adjusted = cv2.convertScaleAbs(rgb_image, alpha=alpha, beta=beta)
cv_imshow(adjusted)
```

```python
# contrast / brightness control

alpha = 1
beta = 20

adjusted = cv2.convertScaleAbs(rgb_image, alpha=alpha, beta=beta)
cv_imshow(adjusted)
```
